```python
import matplotlib.pyplot as plt
import numpy as np
import folium

from tiles import Point, get_images_from_server

server = "http://localhost:8080"
```

```python
left_bottom=Point(55.754718864929245, 12.29589333904291)
top_right=Point(55.773454798193335, 12.342413571948942)

lat = (left_bottom.lat+top_right.lat)/2
lon = (left_bottom.lon+top_right.lon)/2

folium.Map((lat,lon), zoom_start=14)
```

```python
images = get_images_from_server(server, left_bottom, top_right, 14)
fig = plt.figure()
fig.patch.set_facecolor('white')
plt.imshow(np.asarray(images))
plt.show()
```

```python
for zoom in range(7,15):
    print(f"Loading zoom level: {zoom}")
    images = get_images_from_server(server, left_bottom, top_right, zoom)
    fig = plt.figure()
    fig.patch.set_facecolor('white')
    plt.imshow(np.asarray(images))
    plt.show()
```
